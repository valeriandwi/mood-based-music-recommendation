## Train Test Split

```python
df.head(5)
```

```python
kmeans_df = df[[
    "id",
    "uri",
    "track_href",
    "genre",
    "danceability",
    "energy",
    "key",
    "loudness",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo"]]
kmeans_df.head(5)
```

```python
from sklearn.preprocessing import PowerTransformer
import joblib
import numpy as np

def power_transforms(features, df):
  df = df[features]
  for feature in features:
    transformer = PowerTransformer(standardize=True)
    y = np.asanyarray(df[feature])
    y = y.reshape(-1,1)
    transformer.fit(y)
    df["transform_{}".format(feature)] = transformer.transform(y)
    df.drop([feature], axis=1, inplace=True)
    joblib.dump(transformer, "transformer_{}.joblib".format(feature))
  return df
```

```python
transformed_kmeans_df = power_transforms(features=numerical_columns,df=kmeans_df)
transformed_kmeans_df.head()
```

```python
# Get best number of cluster that will be implement

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = {}
silhouette = {}

for k in range(2,10):
    model = KMeans(n_clusters = k, random_state=75)
    model.fit_predict(transformed_kmeans_df)
    inertia[k]= model.inertia_ # Calculate the inertia
    silhouette[k]= silhouette_score(transformed_kmeans_df, model.labels_) # Calculate the silhouette coefficient

fig, axs = plt.subplots(1,2, figsize = (11,4))
axs[0].plot(inertia.keys(), inertia.values(), marker = 'o', lw = 3)
axs[0].set_xlabel('Number of Clusters', fontsize = 12)
axs[0].set_ylabel('Inertia')
axs[0].set_title('The Elbow Method', fontsize = 15)

axs[1].plot(inertia.keys(), silhouette.values(), marker = '*' , lw = 3)
axs[1].set_xlabel('Number of Clusters', fontsize = 12)
axs[1].set_ylabel('Silhouette Coefficient')
axs[1].set_title('The Silhouette Method', fontsize = 15)
plt.show()
```

```python
K = 4

model = KMeans(n_clusters=K, random_state=75)
model.fit(transformed_kmeans_df)

joblib.dump(model, "kmeans_clustering_model.joblib")
```

```python
clusters = model.predict(transformed_kmeans_df)

kmeans_df["mood_segment"] = clusters.astype(str)
kmeans_df.sample(5)
```

```python
#Analytics

result_kmeans_df = pd.merge(
    left=df,
    right=kmeans_df[["id", "mood_segment"]],
    on="id",
    how="inner"
)
result_kmeans_df.head(5)
```

```python
#categorical the result of segments
result_kmeans_df['mood_segment'] = pd.Categorical(result_kmeans_df['mood_segment'], [
    "0", "1", "2", "3"
])
```

```python
sns.countplot(data=result_kmeans_df, x='mood_segment',hue='mood_segment')
plt.title('Count of Mood Segments')
```

```python
def categorical_plot(features, df, segment_feature=None):
    fig, ax = plt.subplots(len(features), 1,figsize=(8,15))
    for i, feature in enumerate(features):
        if segment_feature:
            sns.countplot(data=df, x=segment_feature, hue=feature, ax=ax[i])
        else:
            sns.countplot(data=df, x=feature, ax=ax[i])
    plt.tight_layout()
    plt.show()
```

```python
energy_avg = []
valence_avg = []

moods_segment = ["0","1","2","3"]
for mood in moods_segment :
  energy_avg.append(result_kmeans_df[result_kmeans_df['mood_segment'] == mood]["energy"].mean())
  valence_avg.append(result_kmeans_df[result_kmeans_df['mood_segment'] == mood]["valence"].mean())

plt.figure(figsize=(8, 6))
plt.bar(range(len(energy_avg)), energy_avg)
plt.ylabel('Values')
plt.title('Energy')
plt.xticks(range(len(energy_avg)),moods_segment)  # Optional: labeling x-axis
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(range(len(valence_avg)), valence_avg)
plt.ylabel('Values')
plt.title('Valence')
plt.xticks(range(len(valence_avg)),moods_segment)  # Optional: labeling x-axis
plt.show()
```

0 -> High Energy, High Valence => Happy
1 -> High Energy , Low Valence => Angry
2 -> Low Energy, Low Valence => Sad
3 -> Low Energy, High Valence => Calm
